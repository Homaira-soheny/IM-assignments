```python
import matplotlib.pyplot as plt
import math

#number of items
n = 7
X = [i for i in range(1,n+1,1)]

#constant k for exponential function
k = 4
```

```python
#to connect with the origin append 0 to all the my list 
myLinear = [] 
myQuadratic = []
myCubic = []
myPolynomial = []
myLog = []
for i in (X):
    myLinear.append(i)
    myQuadratic.append(i**2)
    myCubic.append(i**3)
    myPolynomial.append(i**k)
    myLog.append(math.log2(i))

#to connect with the origin    
#X.append(0)
#X.sort()

#Limit the graph to visualize it properly
plt.xlim(0,10)
plt.ylim(0,50)

# Plotting all the curves simultaneously
plt.plot(X,myLinear,'r',label = "Linear", linewidth = 1.0)
plt.plot(X, myQuadratic,'g', label = "Quadratic",linewidth = 1.05)
plt.plot(X,myCubic,'b',label = "Cubic",linewidth = 1.08 )
plt.plot(X, myPolynomial,'m', label="polynomial",linewidth = 1.0)
plt.plot(X, myLog, label="logarithmic",linewidth = 1.0)

# Naming the x-axis, y-axis and the whole graph
plt.xlabel("X (List of items)")
plt.ylabel("Y (Number of operations)")
plt.title("Big-O Notation")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```
